```python
# I executed this code in local, so no need google colab fixes
import cv2
import numpy as np

# Create image with gradient from blue to red
width, height = 256, 256
```

```python
# Create Color.jpeg, a classic uv gradient image 
gradient = np.array([(0, 255 - x, y) for x in range(width) for y in range(height)], dtype=np.uint8).reshape((height, width, 3))
cv2.imshow("Window", gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite("Color.jpeg", gradient)
```

```python
# Rotate upside down
image = np.zeros((height, width, 3), dtype=np.uint8)
cv2.imread("Color.jpeg", image)
image = cv2.flip(image, 0) # 0 means flipping around the x-axis
cv2.imshow("Window", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotate 90 degrees clockwise
image = cv2.imread("Color.jpeg")
image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow("Window", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Turn to grayscale
image = cv2.imread("Color.jpeg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Window", gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
