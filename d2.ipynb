```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
x = np.linspace(0, 8 * np.pi, 1000)

fig, axes = plt.subplots(2, 1, figsize=(8, 12))

fx_series = np.array([(4 / np.pi) * np.sin((2 * n - 1) * x) / (2 * n - 1) for n in range(1, N + 1)])
for fx in fx_series:
    axes[0].plot(x, fx, color="black")

axes[1].plot(x, fx_series.sum(axis=0), color="black")

plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
N = 100
x = np.linspace(0, 8 * np.pi, 1000)

fig, axes = plt.subplots(2, 1, figsize=(8, 12))

fx_series = np.array([2 * ((-1) ** (n + 1)) * np.sin(n * x) / (n) for n in range(1, N + 1)])
for fx in fx_series:
    axes[0].plot(x, fx, color="black")

axes[1].plot(x, fx_series.sum(axis=0), color="black")

plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
